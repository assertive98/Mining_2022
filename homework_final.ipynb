```python
import seaborn as sns흡연상태
sns.countplot(data=health,x="음주여부",hue="성별코드")

#글씨 선명
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")

#그래프 크기
plt.figure(figsize=(15,4))

```

성별과 음주여부를 확인 할 수 있는데 음주여부가 NO인 경우 여성은 2500-3000미만,남성은 1500미만
음주여부가 YES인 경우 여성은 2000 이상, 남성은 3500이상임.
남성이 여성보다 흡연율이 높은 걸 알 수 있다. (결과/인사이트 도출)

```python
sns.pointplot(data=health,x="연령대코드(5세단위)",y="혈색소",hue="음주여부",ci=None)

#글씨 선명
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")

#그래프 크기
plt.figure(figsize=(15,4))

```

연령대코드,혈색소,음주여부에 따라 어떻게 달라지는 그래프를 도출함.
연령대가 올라갈 수록 혈색소가 낮아지는 걸 알 수 있음
음주를 한 사람이 음주를 하지 않은 사람보다 혈색소가 높다는 걸 알 수 있음

```python
sns.pointplot(data=health,x="연령대코드(5세단위)",y="감마지티피",hue="음주여부",ci=None)

#글씨 선명
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")

#그래프 크기
plt.figure(figsize=(15,4))
```

연령대코드,감마지티피,음주여부에 따라 어떻게 달라지는 그래프를 도출함.
감마지티피는 간이 안 좋을때 나타나는 증상 중 하나이며 술을 많이 먹게 되면 높은 수치를 나타낸다. 
도출된 그래프를 보면 연령대가 올라갈 수록 감마지티피가 높아진다.음주여부에 따라 감마지티피가 크게 작용한다는 것을 알 수있다.

```python
sns.scatterplot(data=health,x="수축기혈압",y="이완기혈압",hue="음주여부",marker='D')
```

왼쪽아래에서 오른쪽 위로 선이 그려지는 양의 상관관계가 있어서
음주여부에 대해서 수축기혈압과 이완기혈압에 대해 큰 상관관계를 가지고 있는 것을 알 수 있다. 음주를 하게 되면 혈압에 대해 민감하게 작용되는 것을 보아진다.
