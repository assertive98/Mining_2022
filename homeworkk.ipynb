```python
import pandas as pd
# csv 데이터 가져오기
health = pd.read_csv("/content/drive/My Drive/start/health_data.csv")

display(health.head(10)) # 처음 10행 확인
print(health.info()) #10000데이터 , 31개 Columns
```

```python
import seaborn as sns
sns.countplot(data=health,x="체중(5Kg 단위)",hue="음주여부")

#글씨 선명
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")

#그래프 크기
plt.figure(figsize=(15,4))

##음주를 한다면 체중이 대체로 많이 나가는 경우를 그래프에서 확인할 수 있다. (결과/인사이트 도출)
```

```python
import seaborn as sns
sns.countplot(data=health,x="음주여부",hue="성별코드")

#글씨 선명
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")

#그래프 크기
plt.figure(figsize=(15,4))

#성별과 음주여부를 확인 할 수 있는데 
#음주여부가 NO인 경우 여성은 2500-3000미만,남성은 1500미만
#음주여부가 YES인 경우 여성은 2000 이상, 남성은 3500이상
#남성이 여성보다 흡연율이 높은 걸 알 수 있다. (결과/인사이트 도출)
```

```python
sns.barplot(data=health,x="연령대코드(5세단위)",y="트리글리세라이드",hue="음주여부")

#글씨 선명
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")

plt.figure(figsize=(15,4))

#그래프를 보면 대체적으로 음주를하는 사람들이 트리글리세라이드(중성지방)가 높은 걸 알 수 있다.
#대체적으로 음주를 하는 연령대 코드가 어릴 수록 트리글리세라이드가 높은 걸 알 수 있다.
```

```python
sns.pointplot(data=health,x="연령대코드(5세단위)",y="혈색소",hue="음주여부",ci=None)

#글씨 선명
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")

#그래프 크기
plt.figure(figsize=(15,4))

##연령대가 높을 수록 혈색소가 낮아지는 걸 알 수 있음
##음주를 한 사람이 음주를 하지 않은 사람보다 혈색소가 높다는 걸 알 수 있음
```
